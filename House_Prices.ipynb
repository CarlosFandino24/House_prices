# 2.PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Separate the features from the target variable (if applicable)
X = data.drop('price_sqft_living', axis=1)  # Features

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=X.shape[1])  # Use all features for PCA
X_pca = pca.fit_transform(X_scaled)

# Calculate the explained variance percentage
explained_var_ratio = pca.explained_variance_ratio_
cumulative_var_ratio = explained_var_ratio.cumsum()

# Plot the explained variance percentages
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_var_ratio) + 1), explained_var_ratio * 100, marker='o')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Percentage')
plt.title('Explained Variance Percentage of Individual Components')
plt.xticks(range(1, len(explained_var_ratio) + 1))
plt.show()

# Plot the percentage of total variance captured by all principal components
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_var_ratio) + 1), cumulative_var_ratio * 100, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Percentage of Total Variance Captured')
plt.title('Percentage of Total Variance Captured by Principal Components')
plt.xticks(range(1, len(cumulative_var_ratio) + 1))
plt.show()
```

# 3.Neural Networks Train, Test, Val

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
# Split the data into training and test sets
X_train_val, X_test, y_train_val, y_test = train_test_split(X_pca, data['price_sqft_living'], test_size=0.2, random_state=42)

# Split the training set into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.2, random_state=42)
```

# 4.Neural Networks Cross Validation

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import KFold

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, data['price_sqft_living'], test_size=0.2, random_state=42)

# Define the number of folds for cross-validation
n_splits = 5

# Create an instance of KFold with the desired number of splits
kf = KFold(n_splits=n_splits)

# Initialize lists to store evaluation results
mse_scores = []
mae_scores = []

# Initialize lists to store training and validation loss for each fold
train_losses = []
val_losses = []
```

```python
# Perform cross-validation
for train_index, val_index in kf.split(X_train):
    X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
    y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]  # Use iloc for indexing

    # Define your model
    model2 = keras.Sequential([
        # Input layer
        keras.layers.Dense(64, activation='relu', input_shape=(13,)),
        # Hidden layer
        keras.layers.Dense(64, activation='relu'),
        # Output layer
        keras.layers.Dense(1)
    ])

    # Compile the model
    model2.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_error'])

    # Train the model
    history = model2.fit(X_train_fold, y_train_fold, validation_data=(X_val_fold, y_val_fold), epochs=100, batch_size=32)

    # Get the training and validation loss history for this fold
    training_loss = history.history['loss']
    validation_loss = history.history['val_loss']

    # Append the training and validation loss to the respective lists
    train_losses.append(training_loss)
    val_losses.append(validation_loss)

    # Evaluate the model on validation data
    val_loss, val_mae = model2.evaluate(X_val_fold, y_val_fold)

    # Store the evaluation results
    mse_scores.append(val_loss)
    mae_scores.append(val_mae)

# Convert the lists to arrays for easier manipulation
train_losses = np.array(train_losses)
val_losses = np.array(val_losses)

# Calculate the average training and validation loss across all folds
avg_train_loss = np.mean(train_losses, axis=0)
avg_val_loss = np.mean(val_losses, axis=0)

# Calculate average and standard deviation of evaluation metrics
avg_mse = np.mean(mse_scores)
avg_mae = np.mean(mae_scores)
std_mse = np.std(mse_scores)
std_mae = np.std(mae_scores)

# Print the evaluation results
print("Mean Squared Error (MSE): {:.2f} +/- {:.2f}".format(avg_mse, std_mse))
print("Mean Absolute Error (MAE): {:.2f} +/- {:.2f}".format(avg_mae, std_mae))

# Make predictions on test data or new unseen data
predictions = model2.predict(X_test)
```
